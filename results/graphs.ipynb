```python

def plot_data(config):
  data = config["data"]
  x_axis = config["x_axis"]
  x_label = config["x_label"]
  y_label = config["y_label"]
  title = config["title"]
  group_label_name = config["groups_label"]

  group_labels = sorted(data.keys(), key=lambda x: int(x))  # Ensure order: 100,200,300,500
  library_names = list(next(iter(data.values())).keys())
  colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

  fig, axes = plt.subplots(2, 2, figsize=(13, 10), sharex=True, sharey=True)
  axes = axes.flatten()

  for i, group_label in enumerate(group_labels):
      ax = axes[i]
      libraries = data[group_label]
      for j, lib in enumerate(library_names):
          y_values = libraries[lib]
          ax.plot(x_axis, y_values, marker='o', label=lib, color=colors[j % len(colors)])
      ax.set_title(f"{group_label_name} = {group_label}")
      ax.set_xlabel(x_label)
      ax.set_ylabel(y_label)
      ax.legend()
      ax.grid(True)

  fig.suptitle(title, fontsize=16)
  plt.tight_layout(rect=[0, 0.03, 1, 0.95])
  plt.show()
```
